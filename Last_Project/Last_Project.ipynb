```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
corr = hotel_num_df.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(16, 14))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmin = -1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.savefig("Figures/Fig_correlation.png")
```

```python
from scipy import stats
def TestIndependence(colX,colY):
        X = hotel_num_df[colX].astype(str)
        Y = hotel_num_df[colY].astype(str)
        dfObserved = pd.crosstab(Y,X)
        chi2, p, dof, expected = stats.chi2_contingency(dfObserved.values)
        p = p
        chi2 = chi2
        dof = dof
        return(p)
```

```python
categorical_columns=['hotel', 'lead_time', 'stays_in_weekend_nights',
       'stays_in_week_nights', 'adults', 'children', 'babies', 'meal', 'market_segment', 'distribution_channel',
       'is_repeated_guest', 'previous_cancellations',
       'previous_bookings_not_canceled', 'reserved_room_type',
       'assigned_room_type', 'booking_changes', 'deposit_type', 'agent',
       'company', 'days_in_waiting_list', 'customer_type',
       'required_car_parking_spaces', 'total_of_special_requests',
       'reservation_status']
```

```python
target='is_canceled'
alpha = 0.2
asocs = []
probs = []
```

```python
for var in categorical_columns:
    hotel_num_df[var].fillna('sinInfo',inplace=True)
    hotel_num_df[var] = hotel_num_df[var].astype(str)
    var_classes = hotel_num_df[var].unique().size
    if (var_classes < 100) & (var != target):
        p = TestIndependence(target,var)
        if p<alpha:
            print('Dependence between: ' + str(target) + ' and ' + str(var))
            asocs.append(var)
            probs.append(var)

            plt.figure()
            sns.set(style="whitegrid")
            df_heat = pd.crosstab(hotel_num_df[target],hotel_num_df[var])
            for i in range(df_heat.shape[0]):
                df_heat.iloc[i,:] = df_heat.iloc[i,:]/df_heat.iloc[i,:].sum()
            plt.figure()
            f, ax = plt.subplots(figsize=(5, 5))
            ax = sns.heatmap(df_heat, 
                             annot=False, fmt="d" #, linewidths=.5
                            )
            plt.savefig("Figures/Fig_"+ var +".png")
```

```python
sample_hotel_df = hotel_new_data.sample(n = 8763)
```

```python
data=sample_hotel_df[['hotel', 'stays_in_weekend_nights',
       'stays_in_week_nights', 'adults', 'children', 'babies', 'meal', 'market_segment', 'distribution_channel',
       'is_repeated_guest', 'previous_cancellations',
       'previous_bookings_not_canceled', 'reserved_room_type',
       'assigned_room_type', 'booking_changes', 'deposit_type', 'agent',
       'company', 'customer_type',
       'required_car_parking_spaces', 'total_of_special_requests']]
target = sample_hotel_df['is_canceled']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=250)
rf = rf.fit(X_train, y_train)
```
